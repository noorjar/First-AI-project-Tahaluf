# Resume Cleaning

```python
def clean(text):
    patterns = [
        r'http\S+|www\S+|https\S+',
        r'#\S+\s',
        r'@\S+',
        #r'[!"#$%&\'()*+-.\:;<=>?@[\]^_`{|}~]',
        r'[^\x00-\x7f]',
        r'\s+',
        r' \n ',
        r' \\\/// '
    ]
    for pattern in patterns:
        text = re.sub(pattern, ' ', text)
    return text

df['Cleaned_Text'] = df['Resume_str'].apply(clean)
```

# POSITION_TITLE

```python
def extract_positiontitle(text):
    text = text.strip()  # Remove leading and trailing whitespaces
    if text.upper().startswith('SR'):
        pattern = r'(?i)\bSr\.\s?(?:\w+\s){1}\w+'
    else:
        pattern = r'^([A-Z\s\/\-(),.&\d:]+)\b'
        
    match = re.match(pattern, text)
    if match:
        return match.group(0).strip()
    else:
        return None 

# Apply the function to the 'Cleaned_Text' column of the DataFrame
df['POSITION_TITLE'] = df['Cleaned_Text'].apply(extract_positiontitle)
```

```python
print(df['POSITION_TITLE'].isnull().sum())
null_values_df = df[df['POSITION_TITLE'].isnull()]
null_indices = null_values_df.index.tolist()
print(null_indices)
```

```python
import pandas as pd

most_common_titles = df.groupby('Category')['POSITION_TITLE'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)

# Function to fill missing 'POSITION_TITLE' based on 'Category'
def fill_missing_title(row):
    if pd.isnull(row['POSITION_TITLE']):
        return most_common_titles[row['Category']]
    else:
        return row['POSITION_TITLE']

# Apply the function to fill missing 'POSITION_TITLE'
df['POSITION_TITLE'] = df.apply(fill_missing_title, axis=1)
```

# QUALIFICATIONS

```python
def extract_qualifications(text):
    qualifications = []

    # Define patterns for qualifications
    patterns = [
        r"(certified\s+\w+\s+[\w\s,]+)",
        r"(completed\s+\w+\s+[\w\s,]+)",
        r"(skilled\s+\w+\s+[\w\s,]+)",
        r"(educated\s+\w+\s+[\w\s,]+)",
        r"(member\s+\w+\s+[\w\s,]+)",
        r"(volunteer\s+\w+\s+[\w\s,]+)",
        r"(attended\s+\w+\s+[\w\s,]+)",
        r"(course\s+\w+\s+[\w\s,]+)",
        r"(training\s+\w+\s+[\w\s,]+)",
        r"(skill\s+\w+\s+[\w\s,]+)",
        r"(education\s+\w+\s+[\w\s,]+)",
        r"(qualified\s+\w+\s+[\w\s,]+)",
        r"(experienced\s+\w+\s+[\w\s,]+)",
        r"(trained\s+\w+\s+[\w\s,]+)",
        r"(major\s+in\s+[\w\s,]+)",
    ]

    # Iterate over patterns and search for matches
    for pattern in patterns:
        matches = re.findall(pattern, text, re.IGNORECASE)
        for match in matches:
            qualifications.append(match.strip())

    return qualifications if qualifications else None



df['QUALIFICATIONS'] = df['Cleaned_Text'].apply(extract_qualifications)
```

# Skills

```python
def extract_skills(text):
    # Define a regular expression pattern to match variations of "Skills" case-insensitively
    pattern = re.compile(r'\bSkills\b', re.IGNORECASE)
    # Split the text into sentences
    sentences = re.split(r'[.!?]', text)
    # Initialize a list to store data after each occurrence of "Skills"
    data_list = []

    # Iterate through each sentence to find "Skills" and extract data after it until the end of the sentence
    for sentence in sentences:
        match = pattern.search(sentence)
        if match:
            # Find the index of the match
            skills_index = match.end()
            # Extract data after "Skills" until the end of the sentence
            data_text = sentence[skills_index:].strip()
            if data_text:
                data_list.append(data_text)

    return data_list

df['SKILLS'] = df['Cleaned_Text'].apply(extract_skills)
```

# EDUCATION_MAJOR

```python
def extract_education_major(text):
    # List of education levels
    education_levels = ["BACHELOR", "MASTER", "DIPLOMA", "Master", "master", "Bachelor", "bachelor", "diploma", "Diploma", "Masters", "masters", "Bachelors", "bachelors", "Diplomas", "diplomas", "M.A", "M.a", "B.A", "B.A.", "B.a", "B.S", "B.s", "BSc", "BA", "Ph.D.", "Associates", "BS", "MA", "MSc", "Associate", "MBA", "BSBM", "GED", "AAS", "BSBA", "JD", "M.D.","AS","MBA"]

    # Regular expression patterns for different education major formats
    pattern1 = r'(?i)(?:' + '|'.join(education_levels) + r')\s+of\s+([A-Za-z\s]+)'
    pattern2 = r'(?i)(?:' + '|'.join(education_levels) + r')\s+in\s+([A-Za-z\s]+)'
    pattern3 = r'(?i)(?:' + '|'.join(education_levels) + r')\s+major\s*:\s*([A-Za-z\s]+)'
    pattern4 = r'(?i)(?:' + '|'.join(education_levels) + r')\s+degree\s+in\s*([A-Za-z\s]+)'
    pattern5 = r'(?i)(?:' + '|'.join(education_levels) + r')\s+studies\s+in\s*([A-Za-z\s]+)'

    # Find all matches in the text using different patterns
    matches1 = re.finditer(pattern1, text)
    matches2 = re.finditer(pattern2, text)
    matches3 = re.finditer(pattern3, text)
    matches4 = re.finditer(pattern4, text)
    matches5 = re.finditer(pattern5, text)

    # Extract education majors from different matches
    education_majors1 = [match.group(1).strip() for match in matches1]
    education_majors2 = [match.group(1).strip() for match in matches2]
    education_majors3 = [match.group(1).strip() for match in matches3]
    education_majors4 = [match.group(1).strip() for match in matches4]
    education_majors5 = [match.group(1).strip() for match in matches5]

    # Combine all education majors and remove duplicates
    education_majors = list(set(education_majors1 + education_majors2 + education_majors3 + education_majors4 + education_majors5))

    # Return None if the resulting list is empty
    if not education_majors:
        return None
    
    return education_majors

# Extract education majors from the dataset
df['majors'] = df['Cleaned_Text'].apply(extract_education_major)

```

# EXPERIENCE_LENGTH

```python
def extract_experience(text):
    experience_years = []

    # Define patterns
    patterns = [
        r"(\d+\s*\+?\s*years(?:\s*of\s*experience)?)",
    ]

    # Iterate over patterns and search for matches
    for pattern in patterns:
        matches = re.findall(pattern, text, re.IGNORECASE)
        for match in matches:
            # Extract the numeric part of the match and convert it to an integer
            years_str = re.search(r'\d+', match).group()
            experience_years.append(int(years_str))

    total_experience = sum(experience_years)
    return total_experience if experience_years else None

def extract_experience2(text):
    # Define keywords indicating experience
    experience_keywords = ['experience', 'years', 'yrs']

    # Initialize total experience counter
    total_experience = 0

    # Split text into sentences and iterate over each sentence
    for sentence in text.split('.'):
        # Convert sentence to lowercase for case-insensitive matching
        sentence_lower = sentence.lower()
        
        #========================================================================
        if any(keyword in sentence_lower for keyword in experience_keywords):
            # Extract numbers from the sentence using regex
            numbers = re.findall(r'\b\d+\b', sentence_lower)
            
            # Filter out numbers that are likely indicative of experience
            experience_numbers = [int(num) for num in numbers if 0 < int(num) <= 100]
            
            # If there are numbers indicative of experience, sum them up
            if experience_numbers:
                total_experience += sum(experience_numbers)

    return total_experience if total_experience else None

def combined_extract_experience(text):
    # Apply the first function
    result = extract_experience(text)
    
    # If the first function returns None, apply the second function
    if result is None:
        result = extract_experience2(text)
    
    return result
####################################################################################################
df['experience_length'] = df['Cleaned_Text'].apply(combined_extract_experience)
```

### <span style="color:brown">Q3:

what is the most common skills among all applications?

```python
### Write you code here ###
from collections import Counter

# Assuming df['extracted_data'] contains the text samples
def count_total_skills(text_samples):
    all_skills = []
    for text_list in text_samples:
        if text_list is not None:
            for text in text_list:
                # Split each text by comma and add individual skills to the list
                skills = [skill.strip() for skill in text.split(',')]
                all_skills.extend(skills)
    return all_skills

# Flatten the list of skills from all applications
all_skills = count_total_skills(df['SKILLS'])

# Count the occurrences of each skill
skill_counts = Counter(all_skills)

# Find the most common skills and their counts
most_common_skills = skill_counts.most_common()

# Print the most common skills
print("Most common skills among all applications:")
for skill, count in most_common_skills:
    print(f"- {skill}: {count} occurrences")
```
